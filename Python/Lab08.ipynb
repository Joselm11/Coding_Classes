```python
#Jose Moreno
# modified create account

def get_full_name():
    bad_name = True
    while bad_name:
        name = input("Enter full name: ").strip()
        if " " in name:
            bad_name = False
            return name
        else:
            print("You must enter your full name.")
            
def get_password():
    bad_pwd = True
    while bad_pwd:
        password = input("Enter password:  ").strip()
        digit = False
        cap_letter = False
        for char in password:
            if char.isdigit():
                digit = True
            elif char.isupper():
                cap_letter = True
        if digit == False or cap_letter == False or len(password) < 8:
            print("Password must be 8 characters or more \n" +
                  "with at least one digit and one uppercase letter.")
        else:
            bad_pwd = False
            return password

def get_email():
    """
     making sure email is valid by checking where @ and . is
    """
    bad_email = True
    while bad_email:
        email = input("Enter email address:  ").strip()
        if '@' in email and email.endswith(".com"):
            at_position = -1
            last_dot_position = -1
            
            for i, char in enumerate(email):
                if char == '@':
                    at_position = i
                if char == '.':
                    last_dot_position = i
            
            if at_position != -1 and last_dot_position != -1 and at_position < last_dot_position:
                return email
        else:
            print("Please enter a valid email address")

def get_phone():
    """
    removing non-digit characters and checking if there at 10 digits
    """
    bad_phone = True
    while bad_phone:
        phone = input("Enter a phone number:  ").strip()
        digits_only = ''.join(filter(str.isdigit, phone))
        if len(digits_only) == 10 and digits_only.isdigit():
            return phone
        else:
            print("Please enter a 10 digit phone number. ")

def format_phone(phone):
    """
    formating the phone after input to desired output
    """
    clean_phone = phone.replace('(', '').replace(')', '').replace('-', '').replace(' ', '')
    formatted_phone = f"{clean_phone[:3]}.{clean_phone[3:6]}.{clean_phone[6:]}"
    return formatted_phone
        
def get_first_name(full_name):
    index1 = full_name.find(" ")
    first_name = full_name[:index1]
    return first_name

def main():
    """
    Calling all functions to run and obtain information from user.
    """
    full_name = get_full_name()

    
    password = get_password()


    email = get_email()


    phone = get_phone()

    formatted_phone = format_phone(phone)
    
    first_name = get_first_name(full_name)
    print("Hi " + first_name + ", thanks for creating an account.")
    print(f"We'll text your confirmation code to this number: {formatted_phone}")
    
if __name__ == "__main__":
    main()
```
